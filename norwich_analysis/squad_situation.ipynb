```python
wages.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
players = pd.read_csv("C:/Users/44773/Documents/norwich_players.csv")
```

```python
players.columns
```

```python
wages = wages[wages['Player'].isin(players['Player'])]
```

```python
players = players.merge(wages[['Player', 'Weekly Wage', 'Contract Start', 'Contract End']], how='left', on='Player')
```

```python
players.loc[players['Player'] == 'Lucien Mahovo', 'Contract Start'] = 'Jan 1, 2025'
players.loc[players['Player'] == 'Lucien Mahovo', 'Contract End'] = 'Jun 30, 2028'
players.loc[players['Player'] == 'Lucien Mahovo', 'Weekly Wage'] = 1500
```

```python
players = players[['Player', 'Position', 'Date of birth/Age', 'Weekly Wage', 'Contract Start', 'Contract End']]
```

```python
players['Date of birth/Age'] = players['Date of birth/Age'].str[:-4]
```

```python
players.rename(columns={'Date of birth/Age':'Date of Birth'}, inplace=True)
```

```python
players['Date of Birth'] = (
    players['Date of Birth']
    .str.strip()  # Remove any leading/trailing spaces
    .replace('', pd.NA)  # Handle empty strings properly as NaT
)

players['Date of Birth'] = pd.to_datetime(players['Date of Birth'], format='%b %d, %Y', errors='coerce')
```

```python
players['Contract Start'] = (
    players['Contract Start']
    .str.strip()  # Remove any leading/trailing spaces
    .replace('', pd.NA)  # Handle empty strings properly as NaT
)

players['Contract Start'] = pd.to_datetime(players['Contract Start'], format='%b %d, %Y', errors='coerce')
```

```python
players['Contract End'] = (
    players['Contract End']
    .str.strip()  # Remove any leading/trailing spaces
    .replace('', pd.NA)  # Handle empty strings properly as NaT
)

players['Contract End'] = pd.to_datetime(players['Contract End'], format='%b %d, %Y', errors='coerce')
```

```python
players
```

```python
today = pd.to_datetime(datetime.today().date())

# Assuming your column is already named 'Date of birth':
players['Age'] = players['Date of Birth'].apply(lambda dob: (today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))) if pd.notnull(dob) else pd.NA)
```

```python
players['Contract End Year'] = pd.to_datetime(players['Contract End'], errors='coerce').dt.year
```

```python
players['Years Left'] = players['Contract End Year'] - 2025
```

```python
players
```

```python
def calculate_precise_age(dob):
    if pd.notnull(dob):
        return (today - dob).days / 365.25  # Divide days by average year length
    else:
        return pd.NA

players['Precise Age'] = players['Date of Birth'].apply(calculate_precise_age)
```

```python
# Identify the 'Might Leave' players
players = players[players['Player']!='Lewis Dobbin']
might_leave = ['Marcelino Núñez', 'Borja Sainz', 'Josh Sargent', 'Jack Stacey', 'George Long', 'Kenny McLean', 'Angus Gunn', 'Shane Duffy', 'Onel Hernández', 'Jacob Lungi Sørensen']
players['Might Leave'] = players['Player'].apply(lambda x: 'Might Leave' if x in might_leave else 'Other')

# ===== Plotting =====
plt.figure(figsize=(12, 8))

# Normalize years left for color mapping (Red = expiring, Green = long contract)
norm = plt.Normalize(players['Years Left'].min(), players['Years Left'].max())
cmap_corrected = plt.cm.get_cmap('RdYlGn')  # Red = expiring, Green = max years left

# Plot all players
scatter = plt.scatter(
    players['Precise Age'],
    players['Weekly Wage'],
    c=players['Years Left'],
    cmap=cmap_corrected,
    s=100,
    edgecolor='white',
    alpha=0.9,
    marker='o'
)

# Add labels for 'Might Leave' players
might_leave_players = players[players['Might Leave'] == 'Might Leave']
for _, row in might_leave_players.iterrows():
    if row['Player'] != 'Marcelino Núñez':
        plt.text(
            row['Precise Age'] + 0.1,                # Offset slightly to the right
            row['Weekly Wage'] + 500,               # Offset slightly above
            row['Player'],
            fontsize=9,
            ha='left',
            va='bottom',
            fontweight='bold',
            color='black'
            # backgroundcolor='white'
        )
    else:
        plt.text(
            row['Precise Age'] + 0.1,                # Offset slightly to the right
            row['Weekly Wage'] - 1000,               # Offset slightly above
            row['Player'],
            fontsize=9,
            ha='left',
            va='bottom',
            fontweight='bold',
            color='black'
            # backgroundcolor='white'
        )

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Years Left on Contract')

# ===== Plot Styling =====
plt.title('Age vs. Weekly Wage\nColour Gradient = Years Left on Contract (Red = Expiring)', fontsize=14, fontweight='bold')
plt.xlabel('Precise Age')
plt.ylabel('Weekly Wage (£)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
